```python
data = data.drop('Flow (m¬≥/s)',axis=1)
```

```python
X = data.drop(columns=['Faecal Coliforms (counts/100mL)', 'ùê∏. ùëêùëúùëôùëñ  (counts/100mL)'], axis=1)
y = data[['Faecal Coliforms (counts/100mL)', 'ùê∏. ùëêùëúùëôùëñ  (counts/100mL)']]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=25)
```

```python
X_train.shape, X_valid.shape, y_train.shape, y_valid.shape
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

categorical_cols = X.select_dtypes(include=['object']).columns
numeric_cols  = X.select_dtypes(include=['int64', 'float64']).columns

categorical_cols_indicies = [X.columns.get_loc(col) for col in categorical_cols]
numeric_cols_indicies = [X.columns.get_loc(col) for col in numeric_cols]

ct = ColumnTransformer(
    transformers=[('encoder', OneHotEncoder(handle_unknown='ignore'), categorical_cols_indicies),
                  ('num', StandardScaler(), numeric_cols_indicies)],
    remainder="passthrough"
)

X_train = ct.fit_transform(X_train)
X_valid = ct.transform(X_valid)
```

```python
y_train = y_train['Faecal Coliforms (counts/100mL)']  # N·∫øu b·∫°n ch·ªâ mu·ªën ch·ªçn c·ªôt ƒë·∫ßu ti√™n c·ªßa y_train
# y_train
```

```python
# Use Gradient Boosting
from sklearn.ensemble import GradientBoostingRegressor

rf_model = GradientBoostingRegressor(n_estimators=1000, learning_rate=0.01, max_depth=5, min_samples_leaf=10, min_samples_split=10)
rf_model.fit(X_train, y_train)
```

```python
rf_val_preds = rf_model.predict(X_valid)
```

```python
rf_val_preds[:10], y_valid[: 10]
```

```python
rf_model.score(X_valid, y_valid['Faecal Coliforms (counts/100mL)'])
```
