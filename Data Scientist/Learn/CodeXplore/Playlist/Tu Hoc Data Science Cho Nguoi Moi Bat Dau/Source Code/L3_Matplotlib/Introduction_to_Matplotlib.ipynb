### Pyplot API vs Object-Oriented (OO) API
* Pyplot API &rarr;  Quickly
* OO API &rarr; Advanced

#### Pyplot API

```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x), color='blue', linestyle="dashed", label="sin(x)");
plt.plot(x, np.cos(x), color="red", label="cos(x)");

plt.title("Sin(X) and Cos(X)")
plt.xlabel("Bien X");
plt.ylabel("F(x)");
# plt.xlim([0, 4])
# plt.ylim([-0.75, 0.6])
# plt.axis([0, 4, -0.7, 0.6]) # ~ xlim(), ylim()
# plt.axis("tight")
plt.axis("equal")

plt.legend(); # To display label 
# plt.show() # optional, plt.show() is only required for 'non-interactive' mode such as: command-line, .py script, etc.
```

#### Object-Oriented API

```python
# Step 1. Prepare Data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# Step 2. Setup plot
fig, ax = plt.subplots(figsize=(5, 5)) # Figure size = Width & Height of the plot

# Step 3. Plot the data
ax.plot(x, y, color = "green");
ax.set(title="A simple plot",
       xlabel="x-axis",
       ylabel="y-axis");
```

## 2. Most common types of Matplotlib plots
* line (Biểu đồ đường)
* scatter (Biểu đồ phân tán, rời rạc)
* bar
* hist
* subplots()

### Line

```python
# Create an array
x = np.linspace(1, 10, 100)
x[:5]
```

```python
## OOP API plot a line chart
fig, ax = plt.subplots()
ax.plot(x, x**3);
```

### Scatter

```python
#Pyplot API
plt.scatter(x, np.exp(x)); # y = e^(x)
```

```python
# OO API

# Prepare New Data
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000*rng.rand(100)

fig, ax = plt.subplots()

img1 = ax.scatter(x, y, s=sizes, c=colors, cmap="viridis", alpha=0.3);

fig.colorbar(img1);
```

### Bar
* Vertical
* Horizontal

```python
# Prepare Data from a python dictionary
soft_drink_prices = {"Coke": 10,
                     "Pessi": 12,
                     "Sprite": 8}
fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax.set(title="Bach Hoa Xanh's Soft Drink Prices", ylabel="Price ($)")
```

```python
fig, ax = plt.subplots()
ax.barh(soft_drink_prices.keys(), soft_drink_prices.values());
```

### Histogram

```python
# Prepare Data
np.random.seed(42)
student_height = np.random.normal(170, 10, 250)
```

```python
student_height[:10], len(student_height)
```

```python
# Pyplot API
plt.hist(student_height, bins=30);
```

### Subplots
Figure: hình
* Multiple plots on one figure: https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html

```python
# Option 1:
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot data to each axis
ax1.plot(x, x/2);
ax1.plot(x, x*2);

ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax4.hist(student_height, bins=30);
```

```python
# Option 2:
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot data to each axis
ax[0, 0].plot(x, x/2);
ax[0, 0].plot(x, x*2);

ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax[1, 1].hist(student_height, bins=30);
```
