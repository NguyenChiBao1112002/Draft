## 2. Feature Selections

```python
features = ["LotArea", "YearBuilt", "1stFlrSF", "2ndFlrSF", "FullBath", "BedroomAbvGr", "TotRmsAbvGrd"]
```

## 3. Splitting dataset into X and y

```python
# Nếu dùng toàn bộ Dataset để dùng để train -> dẫn đến Overfitting (Toàn bộ dữ liệu fit vào model)
# -> Khi đưa input có cấu trúc khác vào -> xác suất dữ đoán sai cao -> cần chia ra Test Set -> để kiểm tra độ chính xác
X = data[features]
y = data['SalePrice']
```

```python
X.head()
```

```python
X.info()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

# random_sate ~ seed()
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2,random_state=0)
```

### 4. Training Machine Learning Model

```python
from sklearn.tree import DecisionTreeRegressor

dt_model = DecisionTreeRegressor(random_state=1)
```

```python
# Fit training data into model
dt_model.fit(X_train, y_train) 
```

```python
y_preds = dt_model.predict(X_valid.head())
```

```python
y_preds
```

```python
pd.DataFrame({'y': y_valid.head(), 'y_preds': y_preds})
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(X_train, y_train)
```

```python
rf_val_preds = rf_model.predict(X_valid)
```

```python
rf_val_preds[:5]
```

### Predict with a new input

```python
X_valid.head()
```

### 5. Model Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# y_preds = dt_model.predict(X_valid)
```

```python
mae = mean_absolute_error(y_valid, rf_val_preds)
mse = mean_squared_error(y_valid, rf_val_preds)
rmse = np.sqrt(mse)
r2 = r2_score(y_valid, rf_val_preds)
mape = np.mean(np.abs((y_valid - rf_val_preds)/y_valid)) * 100
```

```python
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R-squared (R²): {r2}")
print(f"Mean Absolute Percentage Error (MAPE): {mape}%")
```
