### Data Imputation (sự buộc tội) (Missing Data Replacement)
Datasets often have missing values, and this can cause problems for machine learning algorithms. It is considered good practice to identify and replace missing values in each column of your dataset prior to performing predictive modeling. This method of missing data replacement is referred to as data imputation.

```python
data_df.info()
```

```python
for col in data_df.columns:
    missing_data = data_df[col].isna().sum()
    missing_percent = (missing_data/len(data_df))*100
    print(f"Column {col}: has {missing_percent}% missing data")
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
sns.heatmap(data=data_df.isna(), cmap="Blues", cbar=False, yticklabels=False);
```

```python
X = data_df.iloc[:,:-1].values # Recommend: Convert from Data Frame to NumPy Array
```

```python
X
```

```python
y = data_df.iloc[:,-1].values
```

```python
y
```

```python
from sklearn.impute import SimpleImputer

# Create as instance of Class SimpleImputer: np.nan is empty value in the dataset
imputer = SimpleImputer(missing_values=np.nan, strategy="mean")

imputer.fit(X[:, 1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

```python
X
```

### Encode Categorical Data 
* Mã hóa các dữ liệu danh mục (dữ liệu không phải Numerical)
 
#### Encode Independent variable (X)

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder="passthrough")
X = ct.fit_transform(X)
```

```python
X
```

#### Encode Dependent Variable (y)

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)
```

```python
y
```

#### Splitting the dataset (X=data, y=output) into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
# Vì train_test_split sẽ lấy ngẫu nhiên, nên đặt seed để giống nhau
np.random.seed(42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_test
```

### Feature Scaling 
* Đưa các feature về cùng một thang đo -> để đảm bảo không có sự bias nào giữa các feature
* **BẮT BUỘC PHẢI** thực hiện sau khi đã Splitting Dataset into Training set and Set Test
* Có 2 kỹ thuật có thể dùng: **Standardization** và **Normalization** 

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, 3:] = sc.fit_transform(X_train[:, 3:])

# Chỉ cần Tranform, sử dụng mean và standard deviation của Train Set.

# Cần đảm bảo không có bất cứ sự tính toán nào dùng dữ liệu Test Set, kể cả tính toán mean và standard deviation
# dù là chỉ dựa vào số liệu chính nó (để đảm bảo tính công bằng để kiểm tra độ chính xác).

# Khi bạn sử dụng các tham số thống kê của Training Set để chuẩn hóa Test Set, bạn đảm bảo rằng Test Set 
# không bị lạm dụng thông tin từ chính nó. Test Set phải được coi là một tập dữ liệu "chưa thấy", 
# và việc chuẩn hóa theo thông số từ Training Set là cách duy nhất để đảm bảo rằng bạn đánh giá mô hình 
# một cách công bằng và chính xác.
X_test[:,3:] = sc.transform(X_test[:,3:])
```
